**English:** This cell redefines the DataFrame with a new, more detailed dataset that will be used for the rest of the analysis.
**Hindi:** Yeh cell DataFrame ko ek naye, zyada detailed dataset ke saath redefine karta hai jo baaki analysis ke liye istemal hoga.

```python
import pandas as pd

data = {
    'Name': ['Amit', 'Neha', 'Ravi', 'Sita', 'Karan', 'Tina', 'Meena', 'Vijay', 'Rohit', 'Asha', None],
    'Department': ['HR', 'IT', 'IT', 'Sales', 'HR', 'Finance', 'Sales', 'IT', 'Finance', 'HR', 'IT'],
    'City': ['Pune', 'Mumbai', 'Pune', 'Delhi', 'Pune', None, 'Delhi', 'Pune', 'Mumbai', 'Delhi', 'Pune'],
    'Salary': [50000, 65000, 60000, 55000, None, 62000, 58000, 61000, 63000, 52000, None],
    'Experience': [2, 5, 3, 4, 1, 6, 2, 3, 5, 4, 2],
    'Email': ['amit@gmail.com', 'neha@yahoo.com', 'ravi@gmail.com', 'sita@outlook.com',
              'karan@gmail.com', 'tina@yahoo.com', 'meena@gmail.com', 'vijay@outlook.com',
              'rohit@gmail.com', 'asha@outlook.com', None]
}

df = pd.DataFrame(data)
```

**English:** Display the new DataFrame.
**Hindi:** Naya DataFrame display karein.

```python
df
```

**English:** Fill missing values in 'Name' with 'unknown', in 'Salary' with the column's mean, and in 'Email' with 'No email'.
**Hindi:** 'Name' mein missing values ko 'unknown' se, 'Salary' ko column ke mean se, aur 'Email' ko 'No email' se bharein.

```python
df['Name'].fillna('unkown', inplace = True)
df['Salary'].fillna(df['Salary'].mean().round(), inplace = True)
df['Email'].fillna('No email', inplace=True)
```

**English:** Display the DataFrame after handling the missing values.
**Hindi:** Missing values handle karne ke baad DataFrame display karein.

```python
df
```

**English:** Group the data by 'Department' and calculate the count and mean of 'Salary' for each department.
**Hindi:** Data ko 'Department' se group karein aur har department ke liye 'Salary' ka count aur mean calculate karein.

```python
df.groupby('Department')['Salary'].agg(['count','mean'])
```

**English:** Extract the email domain from the 'Email' column and create a new 'domain' column.
**Hindi:** 'Email' column se email domain nikalein aur ek naya 'domain' column banayein.

```python
df['domain']=df['Email'].str.extract('@(.*)')
```

**English:** Count the number of employees for each email domain.
**Hindi:** Har email domain ke liye employees ki sankhya ginein.

```python
df['domain'].value_counts().sort_values(ascending=False)
```

**English:** Create a pivot table to show the distribution of email domains across different cities.
**Hindi:** Alag-alag shehron mein email domains ka distribution dikhane ke liye ek pivot table banayein.

```python
df.pivot_table(index='City', columns='domain', values='Department', aggfunc='size', fill_value=0)
```

**English:** Filter the DataFrame to create a new one (`df1`) containing only employees with a salary greater than 60,000.
**Hindi:** Sirf 60,000 se zyada salary wale employees ke liye DataFrame ko filter karke ek naya DataFrame (`df1`) banayein.

```python
df1 = df[df['Salary'] > 60000]
```

**English:** For the high-salary employees (`df1`), group by 'Department' and perform multiple aggregations for salary, experience, and employee count.
**Hindi:** High-salary wale employees (`df1`) ke liye, 'Department' se group karein aur salary, experience, aur employee count ke liye multiple aggregations perform karein.

```python
df1.groupby('Department').agg(
    avg_salary=('Salary','mean'),
    max_salary=('Salary','max'),
    min_salary=('Salary','min'),
    avg_experience=('Experience','mean'),
    total_emp=('Name','count')
).sort_values(by='avg_salary', ascending=False)
```

**English:** Group by 'City', calculate the mean salary, then filter for cities where the mean salary is over 48,000.
**Hindi:** 'City' se group karein, mean salary calculate karein, fir un shehron ko filter karein jahan mean salary 48,000 se zyada hai.

```python
df.groupby('City')['Salary'].mean().loc[lambda y : y > 48000].sort_values(ascending=True)
```

**English:** Group by 'Department' and perform aggregations, then filter the results to show departments with an average salary over 60,000.
**Hindi:** 'Department' se group karein aur aggregations perform karein, fir results ko filter karke un departments ko dikhayein jinka average salary 60,000 se zyada hai.

```python
df.groupby('Department').agg(avg_salary =('Salary','mean'),
                            employee_count = ('Name','count')).loc[lambda y:y['avg_salary']>60000]
```
