**English:** Create a dictionary with sample data. This dictionary contains lists of player names, runs, ages, and cities, including some missing values (`np.nan`, `pd.NA`).
**Hindi:** Sample data ke saath ek dictionary banayein. Is dictionary mein khiladiyon ke naam, run, umar, aur shaharon ki list hai, jismein kuchh missing values (`np.nan`, `pd.NA`) bhi shamil hain.

```python
data ={
    'names':['sachin','virat','dhoni','rohit','pant','rahul',np.nan,'jasprit','yuvraj','surya'],
    'runs':[5676,5461,5434,5460,2333,4554,5665,431,3243,pd.NA],
    'age':[43,np.nan,43,34,23,43,54,34,32,23],
    'city':['mumbai','dilhi','ranchi','mumbai','dilhi',np.nan,'mumbai','ahmadabad','panjab','mumbai']
}
```

**English:** Create a pandas DataFrame from the dictionary.
**Hindi:** Dictionary se ek pandas DataFrame banayein.

```python
df = pd.DataFrame(data)
```

**English:** Use `.iloc` to replace the value at the 7th row (index 6) and 1st column (index 0) with 'sunil'.
**Hindi:** 7th row (index 6) aur 1st column (index 0) par value ko 'sunil' se replace karne ke liye `.iloc` ka upyog karein.

```python
df.iloc[6,0] = 'sunil'
```

**English:** Fill missing values in the 'age' column with the mean of the column. The `inplace=True` argument modifies the DataFrame directly.
**Hindi:** 'age' column mein missing values ko column ke mean se bharein. `inplace=True` argument DataFrame ko seedhe modify karta hai.

```python
df['age'].fillna(int(df['age'].mean()),inplace = True)
```

**English:** Intentionally introduce missing values and inconsistencies for cleaning practice. This code adds NaNs, makes some city names lowercase, and adds extra spaces to the 'Category' column.
**Hindi:** Safai ki practice ke liye jaanbujhkar missing values aur inconsistencies introduce karein. Yah code NaNs add karta hai, kuchh city names ko lowercase banata hai, aur 'Category' column mein extra spaces add karta hai.

```python
for col in ['Sales', 'Profit', 'Discount', 'City', 'Customer Name']:
    df.loc[df.sample(frac=0.05).index, col] = np.nan
df.loc[df['City'].sample(frac=0.03).index, 'City'] = df['City'].str.lower()
df['Category'] = df['Category'].apply(lambda x: ' ' + x + ' ' if np.random.rand() < 0.1 else x)
```
